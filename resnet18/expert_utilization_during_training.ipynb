```python
import wandb
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
api = wandb.Api()
```

```python
def create_Expert_utilization_plot(run_path, filename):
    api = wandb.Api()
    run = api.run(run_path)
    expert_importance = run.history()[['Avg. importance validation layer_1/expert_0', 'Avg. importance validation layer_1/expert_1', 'Avg. importance validation layer_1/expert_2', 'Avg. importance validation layer_1/expert_3']]
    fig, ax = plt.subplots(figsize=[20,6])
    plt.xticks(fontsize=16)
    plt.yticks(fontsize=16)
    plt.vlines(80, 0, 0.5, colors='k')
    plt.vlines(120, 0, 0.5, colors='k')

    ax.plot(expert_importance, linewidth=2) 

    ax.set(xlim=(0, 150), ylim=(0, 0.5))
    ax.legend(['Expert 1', 'Expert 2', 'Expert 3', 'Expert 4'], prop={'size': 20}, loc='upper right')
    plt.xlabel("Epoch", fontsize=24)
    plt.ylabel("Importance", fontsize=24)
    plt.yticks(fontsize=17)
    plt.xticks(fontsize=17)
    plt.savefig(filename)
    return expert_importance
```

```python
create_Expert_utilization_plot('lukas-struppek/final_resnet_18/16aj0dnf', 'expert_utilization_training_imp_1.png')
```
